```python
from PIL import Image
import numpy as np
im = Image.open('./Documents/license-letter/0-new.png').convert('RGBA')
pixels = np.unique(np.array(im.getdata()))
data = np.array(im)
```

```python
r, g, b, a = data[78,55,:][0], data[78,55,:][1], data[78,55,:][2], data[78,55,:][3]
print(r, g, b, a)
new_data = []
round_ratio = 3/194*440 # out of 440 pixels
unit_ratio = 1/194*440
holespacing = 45/194*440
holeheight = 4/194*440
holewidth = 7/194*440
holeradius = 2/194*440
holetop = 4/194*440
circle_center_left = 60/194*440
circle_center_up = 30/194*440
circle_radius = 2/194*440
height, width = int(560*2), int(1760*2)
print(type(a))

# blue background
for i in range(height):
    local = []
    for j in range(width):
        radius = round(round_ratio / 440 * width)
        if i < radius - 1 and j < radius - 1 and (i - radius + 1) ** 2 + (j - radius + 1) ** 2 > radius ** 2:
            local.append([np.uint8(255), np.uint8(255), np.uint8(255), np.uint8(0)])
        elif i < radius - 1 and j > width - radius and (i - radius + 1) ** 2 + (j - width + radius) ** 2 > radius ** 2:
            local.append([np.uint8(255), np.uint8(255), np.uint8(255), np.uint8(0)])
        elif i > height - radius and j < radius - 1 and (i - height + radius) ** 2 + (j - radius + 1) ** 2 > radius ** 2:
            local.append([np.uint8(255), np.uint8(255), np.uint8(255), np.uint8(0)])
        elif i > height - radius and j > width - radius and (i - height + radius) ** 2 + (j - width + radius) ** 2 > radius ** 2:
            local.append([np.uint8(255), np.uint8(255), np.uint8(255), np.uint8(0)])
        else:
            local.append([r, g, b, a])
    new_data.append(local)
new_data = np.array(new_data)

for i in range(height):
    for j in range(width):
        unit = round(1.5 / 440 * width)
        # straight lines.
        if i > radius - 1 and i < height - radius and j > unit and j < unit * 3:
            new_data[i, j, :] = [255, 255, 255, 255]
        elif i > radius - 1 and i < height - radius and j < width - unit and j > width - unit * 3:
            new_data[i, j, :] = [255, 255, 255, 255]
        elif i > unit and i < unit * 3 and j < width - radius and j > radius - 1:
            new_data[i, j, :] = [255, 255, 255, 255]
        elif i < height - unit and i > height - unit * 3 and j < width - radius and j > radius - 1:
            new_data[i, j, :] = [255, 255, 255, 255]
        # curves.
        elif i < radius - 1 + unit and j < radius - 1 + unit and (i - radius + 1 - unit) ** 2 + (j - radius + 1 - unit) ** 2 < (radius - 1) ** 2\
            and (i - radius + 1 - unit) ** 2 + (j - radius + 1 - unit) ** 2 > (radius - 1 - unit * 2) ** 2:
            new_data[i, j, :] = [255, 255, 255, 255]
        elif i < radius - 1 + unit and j > width - radius - unit and (i - radius + 1 - unit) ** 2 + (j - width + radius + unit) ** 2 < (radius - 1) ** 2\
            and (i - radius + 1 - unit) ** 2 + (j - width + radius + unit) ** 2 > (radius - 1 - unit * 2) ** 2:
            new_data[i, j, :] = [255, 255, 255, 255]
        elif i > height - radius - unit and j < radius - 1 + unit and (i - height + radius + unit) ** 2 + (j - radius + 1 - unit) ** 2 < (radius - 1) ** 2\
            and (i - height + radius + unit) ** 2 + (j - radius + 1 - unit) ** 2 > (radius - 1 - unit * 2) ** 2:
            new_data[i, j, :] = [255, 255, 255, 255]
        elif i > height - radius - unit and j > width - radius - unit and (i - height + radius + unit) ** 2 + (j - width + radius + unit) ** 2 < (radius - 1) ** 2\
            and (i - height + radius + unit) ** 2 + (j - width + radius + unit) ** 2 > (radius - 1 - unit * 2) ** 2:
            new_data[i, j, :] = [255, 255, 255, 255]

        # 4 license plate anchor holes.
        # fit to scale
        spacing = round(holespacing / 440 * width) 
        h_height = round(holeheight / 440 * width) 
        h_width = round(holewidth / 440 * width) 
        h_radius = round(holeradius / 440 * width) 
        h_top = round(holetop / 440 * width)
        ccl = round(circle_center_left / 440 * width)
        ccu = round(circle_center_up / 440 * width)
        cr = round(circle_radius / 440 * width)
        # draw rectangles.
        if i > h_top and i < h_top + h_height and j > spacing - 1 and j < spacing + h_width + 1:
            new_data[i, j, :] = [255, 255, 255, 0]
        elif i > h_top and i < h_top + h_height and j < width - spacing + 1 and j > width - spacing - h_width - 1:
            new_data[i, j, :] = [255, 255, 255, 0]
        elif i < height - h_top and i > height - h_top - h_height and j > spacing - 1 and j < spacing + h_width + 1:
            new_data[i, j, :] = [255, 255, 255, 0]
        elif i < height - h_top and i > height - h_top - h_height and j < width - spacing + 1 and j > width - spacing - h_width - 1:
            new_data[i, j, :] = [255, 255, 255, 0]
        # draw semicircles.
        if i > h_top and i < h_top + h_height and j > spacing - h_radius and j < spacing and \
            (i - h_top - h_radius) ** 2 + (j - spacing) ** 2 < h_radius ** 2:
            # top left left side semicircle.
            new_data[i, j, :] = [255, 255, 255, 0]
        elif i > h_top and i < h_top + h_height and j > spacing + h_width and j < spacing + h_width + radius and \
            (i - h_top - h_radius) ** 2 + (j - spacing - h_width) ** 2 < h_radius ** 2:
            # top left right side semicircle.
            new_data[i, j, :] = [255, 255, 255, 0]
        elif i > h_top and i < h_top + h_height and j > width - spacing and j < width - spacing + radius and \
            (i - h_top - h_radius) ** 2 + (j - width + spacing) ** 2 < h_radius ** 2:
            # top right right side semicircle.
            new_data[i, j, :] = [255, 255, 255, 0]
        elif i > h_top and i < h_top + h_height and j > width - spacing - h_width - h_radius and j < width - spacing - h_width and \
            (i - h_top - h_radius) ** 2 + (j - width + spacing+h_width) ** 2 < h_radius ** 2:
            # top right left side semicircle.
            new_data[i, j, :] = [255, 255, 255, 0]
            
        elif i > height - h_top - h_height and i < height - h_top and j > spacing - h_radius and j < spacing and \
            (i - height + h_top + h_radius) ** 2 + (j - spacing) ** 2 < h_radius ** 2:
            # bottom left left side semicircle.
            new_data[i, j, :] = [255, 255, 255, 0]
        elif i > height - h_top - h_height and i < height - h_top and j > spacing + h_width and j < spacing + h_width + radius and \
            (i - height + h_top + h_radius) ** 2 + (j - spacing - h_width) ** 2 < h_radius ** 2:
            # bottom left right side semicircle.
            new_data[i, j, :] = [255, 255, 255, 0]
        elif i > height - h_top - h_height and j > width - spacing and j < width - spacing + radius and \
            (i - height + h_top + h_radius) ** 2 + (j - width + spacing) ** 2 < h_radius ** 2:
            # bottom right right side semicircle.
            new_data[i, j, :] = [255, 255, 255, 0]
        elif i > height - h_top - h_height and j > width - spacing - h_width - h_radius and j < width - spacing - h_width and \
            (i - height + h_top + h_radius) ** 2 + (j - width + spacing+h_width) ** 2 < h_radius ** 2:
            # bottom right right side semicircle.
            new_data[i, j, :] = [255, 255, 255, 0]
        # license plate middle circle.
        elif i > ccu - cr and i < ccu + cr and j > ccl - cr and j < ccl + cr and \
            (i - ccu) ** 2 + (j - ccl) ** 2 < cr ** 2:
            new_data[i, j, :] = [255, 255, 255, 255]
        
new_im = Image.fromarray(new_data)
new_im.show()
```
